# Grids

map plot types to rows and columns, good to create similar plots separated by features

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

PairGrid

```python
#Just the grid
sns.PairGrid(iris)
sns.set()
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter) #plot scatter inside pairplot
sns.set()
```

```python
# Map to upper, lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
sns.set()
```

pairplot

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

Facet Grid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Just the Grid
g = sns.FacetGrid(tips,col='time',row='smoker')
sns.set()
```

```python
g = sns.FacetGrid(tips,col='time',row='smoker')
g = g.map(sns.distplot,'total_bill')
sns.set()
```

```python
g = sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
#Notice how the arguments come after plt.scatter call
g = g.map(plt.scatter,'total_bill','tip').add_legend()
sns.set()
```

JointGrid

```python
g = sns.JointGrid(x='total_bill',y='tip',data=tips)
```

```python
g = sns.JointGrid(x='total_bill',y='tip',data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
