# Regression Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

lmplot() linear model plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
sns.set()
```

```python
# 2 scatter plot and two linear fit
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
sns.set()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

Working with Markers

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],scatter_kws={'s':100})
sns.set()
```

Using a Grid

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex') #sex in diff graph
sns.set()
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
sns.set()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

Aspect and Size

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,height=8)
```
